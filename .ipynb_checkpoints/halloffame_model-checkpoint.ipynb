```python
import numpy as np
import pandas as pd
```

```python
hof = pd.read_csv("hof_data.csv").drop(columns=["Unnamed: 0"])
hof.info()
hof
```

```python
from sklearn.model_selection import train_test_split
import statsmodels.formula.api as smf
```

```python
hof_train, hof_test = train_test_split(hof, test_size=0.3, random_state=88)
hof_train.columns = hof_train.columns.str.replace("%","PCT")
hof_test.columns = hof_test.columns.str.replace("%","PCT")
display(hof_train.head(), hof_test.head())
```

```python
def formula_from_cols_2(df, y):
    return y + ' ~ ' + ' + '.join([col for col in df.columns
                                   if not ((col == 'Players') or (col == 'Player_ID') or (col == 'In_HoF'))])
all_formula = formula_from_cols_2(hof_train, 'In_HoF')
all_formula
```

```python
logreg = smf.logit(formula = all_formula,
                       data = hof_train).fit()
print(logreg.summary())
```
